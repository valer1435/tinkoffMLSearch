```python
import pandas as pd
import numpy as np

df = pd.read_csv("data.csv")
df
```

```python
rank_5 = ["the-beatles", "bob-dylan", "elvis-presley", "the-rolling-stones", "chuck-berry", "jimi-hendrix", "james-brown", "little-richard", ]
rank_4 = ["buddy-holly", "led-zeppelin", "queen", "elton-john", "james-blunt", "billy-joel", "john-lennon"]
rank_3 = ["madonna", "lady-gaga", "michael-jackson"]
def get_rating(x):
    if x["ALink"] in rank_3:
        return 3
    if x["ALink"] in rank_4:
        return 4
    if x["ALink"] in rank_5:
        return 5
    else:
        return 1

df["rating"] = df.apply(get_rating, axis =1)

df[df["rating"] > 0]
```

```python
df.to_csv("data.csv", index=False)
```
